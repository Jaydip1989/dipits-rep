## Finding Pattern and Displaying Count and Complaint type

```python
columns = ['City', 'Complaint Type', 'Unique Key']
df_data_complaint_city_grouped= mydata[columns].groupby(['City', 'Complaint Type'], as_index=False)['Unique Key'].count().rename(columns={'Unique Key' : 'No. of complaint'})
```

```python

df_data_complaint_city_grouped
```

```python
complaint_order=df_data_complaint_city_grouped.sort_values(by='No. of complaint', ascending=False)
```
